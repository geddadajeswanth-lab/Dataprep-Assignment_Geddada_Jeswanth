```python
# Task 1 — Clean the Data 
print('='*30)
print('Task 1 — Clean the Data ')
print('='*30)
# Importing data
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'customer_id': [1, 2, 3, 4, 2, 5],
    'age':         [25, np.nan, 35, np.nan, np.nan, 40],
    'city':        ['Mumbai', 'Delhi', np.nan, 'Mumbai', 'Delhi', np.nan],
    'gender':      ['Male', 'Female', 'Female', 'Male', 'Female', 'Male'],
    'annual_income': [40000, 80000, 60000, 120000, 80000, 95000]
})
```

```python
# Quick analysis to understand the data 
df.info()  # -- to check the data types of each column and non null values
df.isnull().sum()  # to check the no of null values present per column
df.describe()  # for a simple statistics analysis part 
```

```python
# cleaning
# Imputing null values in age column with median of age
median_age = df['age'].median()
df['age'] = df['age'].fillna(median_age)
```

```python
# And city with the mode
city_mode = df['city'].mode()
df['city'] = df['city'].fillna(city_mode[0])
```

```python
# Checking any duplicated rows are present
duplicates = df[df.duplicated(keep = False)].sort_values('age')
print(f'\nDupliated Rows :')
print(duplicates)
# Removing the duplicates
df = df.drop_duplicates()
# check the count of duplicates after the removal
duplicate_count = df.duplicated().sum() 
print(f'\nNo of duplicates present after removal : {duplicate_count}','\n')
```

```python
# Task 2 — Encode Categorical Columns

print('='*30)
print('Task 2 — Encode Categorical Columns')
print('='*30)

# One-Hot coding
df['city'].nunique()
# columns before one hot encoding
print(f'\ncolumns before one hot encoding : {df.shape[1]}')
df = pd.get_dummies(df,columns = ['city'],drop_first=True)
# columns after one hot encoding
print(f'columns after one hot encoding : {df.shape[1]}')
```

```python
# Label - Encoding
# Removing any extra spaces if present 
df['gender'] = df['gender'].astype(str).str.strip()
df['gender'] = df['gender'].map({'Male' : 0, 'Female' : 1})

print(f'\nThe resulting data frame :')
print(df,'\n')
```

```python
# Task 3 — Scale Numerical Features

print('='*30)
print('Task 3 — Scale Numerical Features')
print('='*30)

# Min - Max Scaling
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df[['age', 'annual_income']] = scaler.fit_transform(df[['age', 'annual_income']])

print(f'\nDataframe after max min scaling :')
print(df)

# Standardization
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df[['age', 'annual_income']] = scaler.fit_transform(df[['age', 'annual_income']])

print('\nDataframe after standardization :')
print(df)
```
